```python
'''
Q1 
In a factory, the weights of a product are normally distributed with a mean of 500 grams and a standard deviation of 10 grams.
    1. Calculate the probability density function (PDF) at 490 grams and 510 grams. What do these values represent?
    2. Use the cumulative distribution function (CDF) to find the probability that a randomly chosen product weighs less than 495 grams.
    3. Calculate the CDF at 520 grams and interpret what this probability represents in this context.
'''
# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

# Give Values
mean = 500
std_dev = 10

# Values for specific points
x_490 = 490
x_510 = 510
x_495 = 495
x_520 = 520


# 1. Calculate the PDF at 490 grams and 510 grams
pdf_490 = norm.pdf(x_490, mean, std_dev)
pdf_510 = norm.pdf(x_510, mean, std_dev)

# 2. Calculate the CDF for weights less than 495 grams
cdf_495 = norm.cdf(x_495, mean, std_dev)

# 3. Calculate the CDF for weights less than 520 grams
cdf_520 = norm.cdf(x_520, mean, std_dev)

# Output the results
print("PDF at 490 grams:", pdf_490)
print("PDF at 510 grams:", pdf_510)
print("CDF for weights less than 495 grams:", cdf_495)
print("CDF for weights less than 520 grams:", cdf_520)

# Define a range of x values for plotting the PDF and CDF
x = np.linspace(mean - 4*std_dev, mean + 4*std_dev, 100)

# PDF and CDF values for the entire range
pdf_values = norm.pdf(x, mean, std_dev)
cdf_values = norm.cdf(x, mean, std_dev)


# Plotting
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# Plot PDF
ax[0].plot(x, pdf_values, color='blue', label="PDF (Normal)")
ax[0].set_title("Probability Density Function (PDF) - Normal Distribution")
ax[0].set_xlabel("Weight (grams)")
ax[0].set_ylabel("Density")
ax[0].scatter([x_490, x_510], [pdf_490, pdf_510], color='red', zorder=5)  # Highlight points at 490 and 510 grams
ax[0].legend()

# Plot CDF
ax[1].plot(x, cdf_values, color='green', label="CDF (Normal)")
ax[1].set_title("Cumulative Distribution Function (CDF) - Normal Distribution")
ax[1].set_xlabel("Weight (grams)")
ax[1].set_ylabel("Cumulative Probability")
ax[1].scatter([x_495, x_520], [cdf_495, cdf_520], color='purple', zorder=5)  # Highlight points at 495 and 520 grams
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python

'''
Q2
The height of adult men in a certain country follows a normal distribution with a mean of 175 cm and a standard deviation of 8 cm.
    1. Calculate the probability that a randomly chosen man is taller than 180 cm.
    2. Using the CDF, determine the probability that a randomly chosen mans height lies between 165 cm and 185 cm.
    3. Find the height at which the CDF is equal to 0.95 (i.e., the 95th percentile). What does this height mean in this context?
'''

# Importing Libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Given values
mean_height = 175  # Mean height in cm
std_dev_height = 8  # Standard deviation in cm

# Values for specific points
x_180 = 180  # height in cm
x_165 = 165  # height in cm
x_185 = 185  # height in cm

# 1. Probability that a randomly chosen man is taller than 180 cm
prob_taller_180 = 1 - norm.cdf(x_180, mean_height, std_dev_height)

# 2. Probability that a randomly chosen man's height lies between 165 cm and 185 cm
prob_between_165_185 = norm.cdf(x_185, mean_height, std_dev_height) - norm.cdf(x_165, mean_height, std_dev_height)

# 3. Height at which the CDF is equal to 0.95 (95th percentile)
height_95th_percentile = norm.ppf(0.95, mean_height, std_dev_height)

# Output the results
print("Probability that a randomly chosen man is taller than 180 cm:", prob_taller_180)
print("Probability that a randomly chosen man's height lies between 165 cm and 185 cm:", prob_between_165_185)
print("Height at which the CDF is equal to 0.95 (95th percentile):", height_95th_percentile)

# Define a range of x values for plotting the PDF and CDF
x = np.linspace(mean_height - 4 * std_dev_height, mean_height + 4 * std_dev_height, 100)

# PDF and CDF values for the entire range
pdf_values = norm.pdf(x, mean_height, std_dev_height)
cdf_values = norm.cdf(x, mean_height, std_dev_height)

# Plotting
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# Plot PDF
ax[0].plot(x, pdf_values, color='blue', label="PDF (Normal) for Height")
ax[0].set_title("Probability Density Function (PDF) - Height Distribution")
ax[0].set_xlabel("Height (cm)")
ax[0].set_ylabel("Density")
ax[0].scatter([x_180], [norm.pdf(x_180, mean_height, std_dev_height)], color='red', zorder=5)  # Highlight point at 180 cm
ax[0].legend()

# Plot CDF
ax[1].plot(x, cdf_values, color='green', label="CDF (Normal) for Height")
ax[1].set_title("Cumulative Distribution Function (CDF) - Height Distribution")
ax[1].set_xlabel("Height (cm)")
ax[1].set_ylabel("Cumulative Probability")
ax[1].scatter([x_165, x_185], [norm.cdf(x_165, mean_height, std_dev_height), norm.cdf(x_185, mean_height, std_dev_height)], color='purple', zorder=5)  # Highlight points at 165 cm and 185 cm
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
''' 
Q3
A call center receives an average of 10 calls per hour.
    1. What is the probability that they receive exactly 8 calls in a given hour? (Use the PMF.)
    2. What is the probability of receiving 15 or fewer calls in an hour? (Use the CDF.)
    3. Determine the probability of receiving more than 20 calls in an hour.
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Parameters
lambda_rate = 10  # Average rate of calls per hour
x = np.arange(0, 30)  # Range of possible call counts

# PMF Calculation
pmf = poisson.pmf(x, lambda_rate)

# CDF Calculation
cdf = poisson.cdf(x, lambda_rate)

# Plotting PMF
plt.figure(figsize=(12, 6))

# PMF Plot
plt.subplot(1, 2, 1)
plt.bar(x, pmf, color='blue', alpha=0.7, label='PMF', width=0.5)
plt.title('Probability Mass Function (PMF)')
plt.xlabel('Number of Calls')
plt.ylabel('Probability')
plt.xticks(x)
plt.grid(axis='y')
plt.legend()

# CDF Plot
plt.subplot(1, 2, 2)
plt.plot(x, cdf, color='orange', marker='o', label='CDF')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Number of Calls')
plt.ylabel('Cumulative Probability')
plt.xticks(x)
plt.grid()
plt.legend()

# Show plots
plt.tight_layout()
plt.show()
```

```python
'''
Question 3:
    A call center receives an average of 10 calls per hour.
    1. What is the probability that they receive exactly 8 calls in a given hour? (Use the PMF.)
    2. What is the probability of receiving 15 or fewer calls in an hour? (Use the CDF.)
    3. Determine the probability of receiving more than 20 calls in an hour.
'''
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Given parameters
lambda_rate = 10  # Average rate of calls per hour

# 1. Probability of receiving exactly 8 calls in an hour
prob_exactly_8 = poisson.pmf(8, lambda_rate)

# 2. Probability of receiving 15 or fewer calls in an hour
prob_15_or_fewer = poisson.cdf(15, lambda_rate)

# 3. Probability of receiving more than 20 calls in an hour
prob_more_than_20 = 1 - poisson.cdf(20, lambda_rate)

# Display the results
print("Probability of receiving exactly 8 calls in an hour:", prob_exactly_8)
print("Probability of receiving 15 or fewer calls in an hour:", prob_15_or_fewer)
print("Probability of receiving more than 20 calls in an hour:", prob_more_than_20)

# Define a range of x values for plotting the PMF and CDF
x_values = np.arange(0, 30)

# PMF and CDF values for the range
pmf_values = poisson.pmf(x_values, lambda_rate)
cdf_values = poisson.cdf(x_values, lambda_rate)

# Plotting
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# Plot PMF
ax[0].stem(x_values, pmf_values, basefmt=" ", linefmt='blue', markerfmt="bo", label="PMF (Poisson)")
ax[0].set_title("Probability Mass Function (PMF) - Poisson Distribution")
ax[0].set_xlabel("Number of Calls")
ax[0].set_ylabel("Probability")
ax[0].vlines(8, 0, prob_exactly_8, colors='red', linestyles='dashed', label="P(8 calls)")
ax[0].legend()

# Plot CDF
ax[1].plot(x_values, cdf_values, color='green', label="CDF (Poisson)")
ax[1].set_title("Cumulative Distribution Function (CDF) - Poisson Distribution")
ax[1].set_xlabel("Number of Calls")
ax[1].set_ylabel("Cumulative Probability")
ax[1].fill_between(x_values, cdf_values, where=(x_values <= 15), color="lightgreen", alpha=0.5, label="P(<= 15 calls)")
ax[1].axvline(20, color="purple", linestyle="--", label="P(> 20 calls)")
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
'''
Question 4:
In a certain town, accidents happen at a rate of 3 accidents per day on average.
    1. What is the probability of observing exactly 5 accidents on a given day?
    2. Calculate the probability of observing no more than 2 accidents on a given day.
    3. If an accident occurs once every few hours, is this towns accident rate more likely to follow a Poisson or Normal distribution? Explain your reasoning.
''' 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson

# Given parameters
lambda_rate = 3  # Average rate of accidents per day

# 1. Probability of observing exactly 5 accidents on a given day
prob_exactly_5 = poisson.pmf(5, lambda_rate)

# 2. Probability of observing no more than 2 accidents on a given day
prob_no_more_than_2 = poisson.cdf(2, lambda_rate)

# Display the results
print("Probability of observing exactly 5 accidents on a given day:", prob_exactly_5)
print("Probability of observing no more than 2 accidents on a given day:", prob_no_more_than_2)

# Define a range of x values for plotting the PMF and CDF
x_values = np.arange(0, 15)  # Possible range of accident counts

# PMF and CDF values for the range
pmf_values = poisson.pmf(x_values, lambda_rate)
cdf_values = poisson.cdf(x_values, lambda_rate)

# Plotting
fig, ax = plt.subplots(2, 1, figsize=(10, 10))

# Plot PMF
ax[0].stem(x_values, pmf_values, basefmt=" ", linefmt='blue', markerfmt="bo", label="PMF (Poisson)")
ax[0].set_title("Probability Mass Function (PMF) - Poisson Distribution")
ax[0].set_xlabel("Number of Accidents")
ax[0].set_ylabel("Probability")
ax[0].vlines(5, 0, prob_exactly_5, colors='red', linestyles='dashed', label="P(5 accidents)")
ax[0].legend()

# Plot CDF
ax[1].plot(x_values, cdf_values, color='green', label="CDF (Poisson)")
ax[1].set_title("Cumulative Distribution Function (CDF) - Poisson Distribution")
ax[1].set_xlabel("Number of Accidents")
ax[1].set_ylabel("Cumulative Probability")
ax[1].fill_between(x_values, cdf_values, where=(x_values <= 2), color="lightgreen", alpha=0.5, label="P(<= 2 accidents)")
ax[1].legend()

plt.tight_layout()
plt.show()
```

```python
'''
Question 5:
In a school cafeteria, students arrive randomly at a rate of 5 students per minute. The time between arrivals is known to follow an exponential distribution (a related form of the Poisson distribution).
    1. What is the probability that exactly 3 students arrive in the next minute?
    2. Using the CDF, calculate the probability that at least 10 students arrive within the next 2 minutes.
    3. If the waiting times between arrivals are normally distributed with a mean of 12 seconds and a standard deviation of 5 seconds, what is the probability that the waiting time for the next student exceeds 20 seconds?
''' 

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import poisson, norm

# Given parameters
arrival_rate_per_minute = 5  # Average rate of students per minute
waiting_time_mean = 12       # Mean waiting time in seconds
waiting_time_std_dev = 5     # Standard deviation of waiting time in seconds

# 1. Probability that exactly 3 students arrive in the next minute
prob_exactly_3 = poisson.pmf(3, arrival_rate_per_minute)

# 2. Probability that at least 10 students arrive within the next 2 minutes
arrival_rate_per_2_minutes = 2 * arrival_rate_per_minute  # Arrival rate over 2 minutes
prob_10_or_more = 1 - poisson.cdf(9, arrival_rate_per_2_minutes)

# 3. Probability that the waiting time exceeds 20 seconds
prob_wait_time_exceeds_20 = 1 - norm.cdf(20, waiting_time_mean, waiting_time_std_dev)

# Display the results
print("Probability of exactly 3 students arriving in the next minute:", prob_exactly_3)
print("Probability of at least 10 students arriving in the next 2 minutes:", prob_10_or_more)
print("Probability that the waiting time exceeds 20 seconds:", prob_wait_time_exceeds_20)

# Define a range of x values for plotting the PMF and CDF
x_values_poisson = np.arange(0, 20)  # Possible range of student counts for Poisson
x_values_normal = np.linspace(0, 30, 100)  # Range of waiting times for Normal distribution

# PMF and CDF values for the Poisson distribution
pmf_values_poisson = poisson.pmf(x_values_poisson, arrival_rate_per_minute)
cdf_values_poisson_2min = poisson.cdf(x_values_poisson, arrival_rate_per_2_minutes)

# PDF for Normal distribution of waiting times
pdf_values_normal = norm.pdf(x_values_normal, waiting_time_mean, waiting_time_std_dev)

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

# Plot PMF for Poisson distribution (students per minute)
ax[0].stem(x_values_poisson, pmf_values_poisson, basefmt=" ", linefmt='blue', markerfmt="bo", label="PMF (Poisson, =5)")
ax[0].set_title("Probability Mass Function (PMF) - Poisson Distribution")
ax[0].set_xlabel("Number of Student Arrivals (1 minute)")
ax[0].set_ylabel("Probability")
ax[0].vlines(3, 0, prob_exactly_3, colors='red', linestyles='dashed', label="P(3 students)")
ax[0].legend()

# Plot CDF for Poisson distribution (students in 2 minutes)
ax[1].plot(x_values_poisson, cdf_values_poisson_2min, color='green', label="CDF (Poisson, =10)")
ax[1].set_title("Cumulative Distribution Function (CDF) - Poisson Distribution")
ax[1].set_xlabel("Number of Student Arrivals (2 minutes)")
ax[1].set_ylabel("Cumulative Probability")
ax[1].fill_between(x_values_poisson, cdf_values_poisson_2min, where=(x_values_poisson >= 10), color="lightgreen", alpha=0.5, label="P(>= 10 students)")
ax[1].legend()

# Plot PDF for Normal distribution of waiting times
ax[2].plot(x_values_normal, pdf_values_normal, color='purple', label="PDF (Normal)")
ax[2].set_title("Probability Density Function (PDF) - Normal Distribution of Waiting Times")
ax[2].set_xlabel("Waiting Time (seconds)")
ax[2].set_ylabel("Density")
ax[2].fill_between(x_values_normal, pdf_values_normal, where=(x_values_normal >= 20), color="lightcoral", alpha=0.5, label="P(Waiting Time > 20s)")
ax[2].legend()

plt.tight_layout()
plt.show()
```

```python
'''
Question 6:
You are analyzing customer purchase amounts at an e-commerce store and find that these amounts roughly follow a normal distribution with a mean of $75 and a standard deviation of $15.
    1. Find the probability that a randomly selected customer spends more than $100.
    2. Calculate the probability that a customer spends between $60 and $90.
    3. Assume that the store receives 20 purchases per hour on average. What is the probability of receiving more than 25 purchases in an hour?
'''

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, poisson

# Given parameters
purchase_mean = 75       # Mean purchase amount in dollars
purchase_std_dev = 15    # Standard deviation of purchase amount in dollars
average_purchases_per_hour = 20  # Average purchases per hour

# 1. Probability that a customer spends more than $100
prob_spend_more_than_100 = 1 - norm.cdf(100, purchase_mean, purchase_std_dev)

# 2. Probability that a customer spends between $60 and $90
prob_spend_between_60_and_90 = norm.cdf(90, purchase_mean, purchase_std_dev) - norm.cdf(60, purchase_mean, purchase_std_dev)

# 3. Probability of receiving more than 25 purchases in an hour
prob_more_than_25_purchases = 1 - poisson.cdf(25, average_purchases_per_hour)

# Display the results
print("Probability that a customer spends more than $100:", prob_spend_more_than_100)
print("Probability that a customer spends between $60 and $90:", prob_spend_between_60_and_90)
print("Probability of receiving more than 25 purchases in an hour:", prob_more_than_25_purchases)

# Define a range of x values for plotting the PDF and CDF
x_values_normal = np.linspace(30, 120, 100)  # Range for purchase amounts in dollars
x_values_poisson = np.arange(0, 40)          # Possible range of purchase counts per hour

# PDF for Normal distribution of purchase amounts
pdf_values_normal = norm.pdf(x_values_normal, purchase_mean, purchase_std_dev)

# PMF and CDF values for the Poisson distribution (purchases per hour)
pmf_values_poisson = poisson.pmf(x_values_poisson, average_purchases_per_hour)
cdf_values_poisson = poisson.cdf(x_values_poisson, average_purchases_per_hour)

# Plotting
fig, ax = plt.subplots(3, 1, figsize=(10, 15))

# Plot PDF for Normal distribution of purchase amounts
ax[0].plot(x_values_normal, pdf_values_normal, color='purple', label="PDF (Normal)")
ax[0].set_title("Probability Density Function (PDF) - Normal Distribution of Purchase Amounts")
ax[0].set_xlabel("Purchase Amount ($)")
ax[0].set_ylabel("Density")
ax[0].fill_between(x_values_normal, pdf_values_normal, where=(x_values_normal >= 100), color="lightcoral", alpha=0.5, label="P(Spend > $100)")
ax[0].fill_between(x_values_normal, pdf_values_normal, where=((x_values_normal >= 60) & (x_values_normal <= 90)), color="lightblue", alpha=0.5, label="P($60 < Spend < $90)")
ax[0].legend()

# Plot PMF for Poisson distribution (purchases per hour)
ax[1].stem(x_values_poisson, pmf_values_poisson, basefmt=" ", linefmt='blue', markerfmt="bo", label="PMF (Poisson, =20)")
ax[1].set_title("Probability Mass Function (PMF) - Poisson Distribution of Purchases per Hour")
ax[1].set_xlabel("Number of Purchases")
ax[1].set_ylabel("Probability")
ax[1].legend()

# Plot CDF for Poisson distribution (purchases per hour) with area of interest
ax[2].plot(x_values_poisson, cdf_values_poisson, color='green', label="CDF (Poisson, =20)")
ax[2].set_title("Cumulative Distribution Function (CDF) - Poisson Distribution of Purchases per Hour")
ax[2].set_xlabel("Number of Purchases")
ax[2].set_ylabel("Cumulative Probability")
ax[2].fill_between(x_values_poisson, cdf_values_poisson, where=(x_values_poisson > 25), color="lightgreen", alpha=0.5, label="P(> 25 purchases)")
ax[2].legend()

plt.tight_layout()
plt.show()
```
