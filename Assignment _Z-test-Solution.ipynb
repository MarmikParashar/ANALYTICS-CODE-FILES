```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Defining a Function for Calculating the Z-Test Value
def Z_test(sample_mean,population_mean,population_std,sample_size,alpha,plt_title):
    # Step 2: Calculate Z-Test
    z_test = (sample_mean - population_mean) / (population_std / np.sqrt(sample_size))

    # Step 3: Determine critical Z-values for a two-tailed test at a 0.01 significance level
    z_critical = stats.norm.ppf(1 - alpha / 2)

    # Step 4: Compare Z-score with critical Z-value to make a decision
    reject_null = abs(z_test ) > z_critical

    # Step 5: Visualization
    # Standard normal distribution for visualization
    x = np.linspace(-4, 4, 1000)
    y = stats.norm.pdf(x, 0, 1)

    plt.figure(figsize=(10, 6))
    plt.plot(x, y, label="Standard Normal Distribution")
    plt.fill_between(x, y, where=(x <= -z_critical) | (x >= z_critical), color="red", alpha=0.3, label="Rejection Region")
    plt.axvline(z_test , color="blue", linestyle="--", label=f"Z-Test = {z_test :.2f}")
    plt.axvline(-z_critical, color="black", linestyle="--", label=f"Z-Critical = Â±{z_critical:.2f}")
    plt.axvline(z_critical, color="black", linestyle="--")
    plt.xlabel("Z")
    plt.ylabel("Probability Density")
    plt.title(plt_title)
    plt.legend()
    plt.show()

    # Display results
    print("Z-Test:", z_test )
    print("Critical Z-Value:", z_critical)
    print("Reject Null Hypothesis:", reject_null)
    if(reject_null):
        print("There is sufficient evidence to reject the null hypothesis")
    else:
        print("There is insufficient evidence to reject the null hypothesis")
```
