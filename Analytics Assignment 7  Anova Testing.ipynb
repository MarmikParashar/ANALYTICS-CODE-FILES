1. Study Hours and Test Scores
 Columns: Study_Hours (Categorical: "<5", "5-10", ">10"), Test_Score (Continuous)
 Data Simulation: Use np.random.normal to create different mean scores for each
study hour group.

```python
import numpy as np
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Create Synthetic Dataset
np.random.seed(0)
data = pd.DataFrame({
    'Study_Hours': np.repeat(['<5', '5-10', '>10'], 30),
    'Test_Score': np.concatenate([
        np.random.normal(60, 10, 30),  # "<5" hours group
        np.random.normal(70, 10, 30),  # "5-10" hours group
        np.random.normal(80, 10, 30)   # ">10" hours group
    ])
})

# Step 2: Perform ANOVA Test
groups = [data['Test_Score'][data['Study_Hours'] == hour] for hour in data['Study_Hours'].unique()]
anova_result = stats.f_oneway(*groups)

# Step 3: Visualization
plt.figure(figsize=(8, 6))
sns.boxplot(x='Study_Hours', y='Test_Score', data=data, palette="Set2")
plt.title('Test Scores by Study Hours')
plt.xlabel('Study Hours')
plt.ylabel('Test Score')
plt.show()

# Step 4: Conclusion
alpha = 0.05
if anova_result.pvalue < alpha:
    conclusion = "Reject the null hypothesis. There is a significant difference in test scores between study hour groups."
else:
    conclusion = "Fail to reject the null hypothesis. No significant difference in test scores between study hour groups."

print(f"ANOVA Result: F={anova_result.statistic:.2f}, p-value={anova_result.pvalue:.20f}")
print("Conclusion:", conclusion)
```

## Factorial ANOVA Test

```python
import pandas as pd
import numpy as np
from scipy import stats
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm

# Set a random seed for reproducibility
np.random.seed(42)

def perform_two_way_anova(data, formula, factor1, factor2, response):
    # Perform the two-way ANOVA
    model = ols(formula, data=data).fit()
    anova_results = anova_lm(model)
    
    # Print ANOVA results
    print(anova_results)
    
    # Visualization: Heatmap
    pivot_table = data.pivot_table(response, factor1, factor2, aggfunc='mean')
    plt.figure(figsize=(8, 6))
    sns.heatmap(pivot_table, annot=True, cmap='YlGnBu', fmt='.1f', linewidths=0.5)
    plt.title(f'Heatmap of {response} by {factor1} and {factor2}')
    plt.xlabel(factor2)
    plt.ylabel(factor1)
    plt.show()
    
    # Visualization: Boxplot
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=factor1, y=response, hue=factor2, data=data, palette="Set2")
    plt.title(f'{response} Distribution by {factor1} and {factor2}')
    plt.xlabel(factor1)
    plt.ylabel(response)
    plt.legend(title=factor2)
    plt.show()
```

```python
# Question 1: Sales by Region and Season
def question_1():
    data = {
        'Region': np.repeat(['North', 'South', 'East', 'West'], 50),
        'Season': np.tile(np.repeat(['Winter', 'Summer'], 25), 4),
        'Sales': np.random.normal(100, 10, 200) + np.tile(np.repeat([5, -5], 25), 4)
    }
    df = pd.DataFrame(data)
    perform_two_way_anova(df, 'Sales ~ C(Region) * C(Season)', 'Region', 'Season', 'Sales')

question_1()
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from statsmodels.stats.anova import AnovaRM

# Function to perform Repeated Measures ANOVA
def repeated_measures_anova(data, subject_col, time_col, measure_col, title="Repeated Measures ANOVA"):
    # Create the repeated measures ANOVA model
    rm_anova = AnovaRM(data, measure_col, subject_col, within=[time_col]).fit()
    print(rm_anova)

    # Plot lineplot to visualize trends over time
    plt.figure(figsize=(10, 6))
    sns.lineplot(x=time_col, y=measure_col, data=data, ci=None, marker='o', hue=time_col, palette="Set2")
    plt.title(f'{title} - Line Plot')
    plt.xlabel(time_col)
    plt.ylabel(measure_col)
    plt.show()

    # Plot boxplot to show distribution at each time point
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=time_col, y=measure_col, data=data, palette="Set2")
    plt.title(f'{title} - Box Plot')
    plt.xlabel(time_col)
    plt.ylabel(measure_col)
    plt.show()

# Example Datasets for each question:
```
